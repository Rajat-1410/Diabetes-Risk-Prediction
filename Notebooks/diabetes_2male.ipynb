```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings 
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
import sqlite3

conn = sqlite3.connect("../Data/diabetes_project.db")

DATA = pd.read_sql("SELECT * FROM coded_data", conn)
```

```python
DATA['GENDER'].value_counts()
```

```python
DATA.columns=DATA.columns.str.strip()
```

```python
DATA.head()
```

```python
Data_female=DATA[DATA['GENDER']!=0]
```

```python
Data_male=DATA[DATA['GENDER']!=1]
```

```python
Data_female
```

```python
Data_female.drop('GENDER',axis=1,inplace=True)
```

```python
from lifelines import CoxPHFitter
cox_ph=CoxPHFitter()
cox_female=cox_ph.fit(Data_female,'AGE','SUBJECT TYPE')
cox_female.print_summary()
```

```python
Data_male.drop('GENDER',axis=1,inplace=True)
```

```python
cox_male=cox_ph.fit(Data_male,'AGE','SUBJECT TYPE')
cox_male.print_summary()
```

```python
import joblib

joblib.dump(cox_male, "../Models/cox_male.pkl")
joblib.dump(cox_female, "../Models/cox_female.pkl")
```
